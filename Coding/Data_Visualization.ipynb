**Data Visualization**

**Box Plot Distribution of the Target Variable 'Price'**


```python
# Creating boxplots to visualize the distribution of the target variable 'Price' across different levels of attributes

# importing Libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Housing.csv')

# Define the columns to analyze
columns = ['bedrooms', 'bathrooms', 'stories', 'parking' , 'mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']

# Set the color palette for the boxplots
color = 'deep'

# Create a figure with a specific size
plt.figure(figsize=(15, 30))

# Iterate over each column
for i, num in enumerate(columns):
    # Create a subplot for each column
    ax = plt.subplot(7, 3, i + 1)

    # Create a boxplot with the categorical column on the x-axis and the price on the y-axis
    sns.boxplot(x=data[num], y=data['price'], palette=color)

    # Format the y-axis labels to be in plain style
    plt.ticklabel_format(style='plain', axis='y')

    # Set the title for each subplot
    plt.title(num, fontsize=14, fontweight='bold')

    # Set the font size of the x-axis tick labels
    plt.xticks(fontsize=14)

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

**Average Price and the distribution using Bar plot**

```python
# Calculate the average price
average_price = data['price'].mean()

# Create a data frame of houses above and below average price
above_avg = data[data['price'] > average_price]
below_avg = data[data['price'] <= average_price]

# Print the average price of the House and the count of Houses above and below the average price
print("Average Price of the House:", round(average_price,2))
print("Houses above average price:", len(above_avg))
print("Houses below or equal to average price:", len(below_avg))
```

```python
# Analysis of Average Price Variation for Above-Average Houses by Categorical Variables

# Define the variables to analyze
cat = ['bedrooms', 'bathrooms', 'stories', 'mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'parking', 'prefarea', 'furnishingstatus']

# Create a figure with a specific size
plt.figure(figsize=(12, 35))

# Set the super title for the plot
plt.suptitle('Average price of above-average homes (> 4766729) \n', fontsize=16, y=0.99)

# Iterate over each categorical variable
for i, d in enumerate(cat):
    # Create a subplot for each variable
    ax = plt.subplot(11, 2, i + 1)

    # Create a bar plot with the categorical variable on the x-axis and the average price on the y-axis
    sns.barplot(x=above_avg[d], y=above_avg['price'], errorbar=None , palette=color)

    # Format the y-axis labels to be in plain style
    plt.ticklabel_format(style='plain', axis='y')

    # Set the font size of the x-axis tick labels
    plt.xticks(fontsize=14)

    # Add labels to the bars with the average price
    for n in ax.containers:
        ax.bar_label(n, fmt='%.2f', label_type='center', fontsize=14, rotation=90)

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Analysis of Average Price Variation for Below-Average Houses by Categorical Variables

# Define the variables to analyze
cat = ['bedrooms', 'bathrooms', 'stories', 'mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'parking', 'prefarea', 'furnishingstatus']

# Create a figure with a specific size
plt.figure(figsize=(12, 35))

# Set the super title for the plot
plt.suptitle('Average price of below-average homes (<= 4766729) \n', fontsize=16, y=0.99)

# Iterate over each categorical variable
for i, d in enumerate(cat):
    # Create a subplot for each variable
    ax = plt.subplot(11, 4, i + 1)

    # Create a bar plot with the categorical variable on the x-axis and the average price on the y-axis
    sns.barplot(x=below_avg[d], y=below_avg['price'], errorbar=None , palette=color)

    # Format the y-axis labels to be in plain style
    plt.ticklabel_format(style='plain', axis='y')

    # Set the font size of the x-axis tick labels
    plt.xticks(fontsize=14)

    # Add labels to the bars with the average price
    for n in ax.containers:
        ax.bar_label(n, fmt='%.2f', label_type='center', rotation=90, fontsize=14)

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
