**Univariant Analysis**

```python
# Select the numerical attributes for univariate analysis
numerical_attributes = ['price', 'area', 'bedrooms', 'bathrooms', 'stories']

# Set the figure size
plt.figure(figsize=(15, 10))

# Iterate over each numerical attribute
for i, attribute in enumerate(numerical_attributes):
    ax = plt.subplot(2, 3, i+1)  # Create subplots
    sns.histplot(data[attribute], kde=True)  # Plot histogram with kernel density estimation
    plt.title(attribute)  # Set title for each attribute
    plt.xlabel('')  # Remove x-axis label

    # Calculate basic statistics
    column_mean = data[attribute].mean()
    column_median = data[attribute].median()
    column_std = data[attribute].std()
    column_skewness = data[attribute].skew()

    # Display statistics
    stats_text = f"Mean: {column_mean:.2f}\nMedian: {column_median:.2f}\nStd Dev: {column_std:.2f}\nSkewness: {column_skewness:.2f}"
    plt.text(0.95, 0.9, stats_text, transform=ax.transAxes, ha='right', va='top')  # Display statistics

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

**Bivariant Analysis**

```python
# Select the numerical attributes for bivariate analysis
numerical_attributes = ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']

# Create a scatter plot matrix
sns.set(style="ticks")
sns.pairplot(data[numerical_attributes])

# Show the plot
plt.show()
```

**Categorical Feature Analysis**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("Housing.csv")

# Categorical features for analysis
categorical_features = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'parking', 'prefarea', 'furnishingstatus']

# Plotting count plots for each categorical feature
plt.figure(figsize=(15, 10))
for i, feature in enumerate(categorical_features, 1):
    plt.subplot(2, 4, i)
    counts = data[feature].value_counts()
    counts.plot(kind='bar')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(f'{feature.capitalize()} Count')

    # Add count labels on the bars
    for j, count in enumerate(counts):
        plt.text(j, count, str(count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

**Outliers**

```python
# Outliers Detection
import pandas as pd
import numpy as np

# Load the dataset
data = pd.read_csv('Housing.csv')
# Select numerical columns
numerical_columns = data.select_dtypes(include=np.number)

# Calculate the upper and lower bounds for outliers using the z-score method
z_scores = (numerical_columns - numerical_columns.mean()) / numerical_columns.std()
outliers = (z_scores > 3) | (z_scores < -3)

# Calculate the count and percentage of outliers for each column
outlier_count = outliers.sum()
outlier_percentage = (outlier_count / len(data)) * 100

# Create a table to display the results
result_table = pd.DataFrame({'Outlier Count': outlier_count, 'Outlier Percentage': outlier_percentage})
print(result_table)
```

```python
# Detection of outliers using boxplot

import matplotlib.pyplot as plt
import seaborn as sns

# Define the function to create a boxplot for a given column
def boxplot(column):
    plt.subplot(2, 3, i+1)  # Create a subplot
    sns.boxplot(data=data, x=data[f"{column}"])  # Create the boxplot
    plt.title(f"Boxplot of {column}")  # Set the title
    plt.tight_layout()  # Adjust subplot spacing

# Create the subplots for outlier detection
plt.figure(figsize=(12, 8))  # Set the figure size

# Iterate over each numerical column
for i, column in enumerate(["price", "bedrooms", "bathrooms", "parking", "area", "stories"]):
    boxplot(column)  # Create the boxplot for the current column

# Display the subplots
plt.show()
```
